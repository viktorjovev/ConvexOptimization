```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LogisticRegression
```

```python
data = pd.read_excel("/content/advertising.xlsx")
```

```python
data
```

```python
numerical_features = data[['Daily Time Spent on Site','Age','Area Income', 'Daily Internet Usage','Male', 'Clicked on Ad']]
numerical_features
```

```python
regression = LogisticRegression()
regression.fit(numerical_features.drop(columns=['Clicked on Ad']),numerical_features['Clicked on Ad'])
print("Regression Coefficient: ", regression.coef_)
print("Regression Intercept: ", regression.intercept_)
```

# Encoding the textual features

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

encoder.fit(data['Ad Topic Line'])
data['Ad Topic Line'] = encoder.transform(data['Ad Topic Line'])
```

```python
encoder.fit(data['City'])
data['City'] = encoder.transform(data['City'])
```

```python
encoder.fit(data['Country'])
data['Country'] = encoder.transform(data['Country'])
```

```python
data
```

#Testing with the encoded textual features

```python
regression = LogisticRegression()
regression.fit(data.drop(columns=['Clicked on Ad', 'Timestamp']),data['Clicked on Ad'])
print("Regression Coefficient: ", regression.coef_)
print("Regression Intercept: ", regression.intercept_)
```
